```python
import numpy as np
import pylab
import math
import matplotlib.pyplot as plt
from scipy.stats import expon
%matplotlib inline

# Поддержка русскоязычных надписей.
pylab.rc('font',**{'family':'verdana'})
```

```python
def showEstimation(k) :
    maxSize = 10000
    estimation = np.array([])
    # grid - список 1..N
    grid = np.array([])

    # Генерируем выборку из экспоненциального распределения с параметром 1.
    sample = expon.rvs(loc=1, size=maxSize)
    
    factorial = math.factorial(k)
    # Для каждого n <= N считаем значение всех оценок на выборке x[0]..x[n]
    for sampleSize in range(1, maxSize, 1):
        # Считаем оценку.
        mometK = list(map(lambda x : x**k, sample[:sampleSize]))
        value = (factorial / np.mean(mometK)) ** (1/k)
        grid = np.append(grid, sampleSize)
        estimation = np.append(estimation, abs(value - 1))

    # Строим график зависимости оценки от n.
    plt.plot(grid, estimation, label="k = " + str(k))
```

```python
plt.figure(figsize=(20, 20))
for k in range(4, 15, 2) :
    showEstimation(k)
plt.ylim(0, 0.4)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 20))
for k in range(5, 15, 2) :
    showEstimation(k)
plt.ylim(0, 0.4)
plt.legend()
plt.show()
```
